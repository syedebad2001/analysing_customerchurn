# lets have a look at average utlization ratios by ordinal variables such as gender, marital status, income, and education level


```python
# education (uper kahin adjust karo)
data[data['Avg_Utilization_Ratio']==0].value_counts().shape[0]
# we see that 20% of the data is 0 so to have better results lets look at people with greater than 0 avg util
```

```python
# gender
print(data[data['Avg_Utilization_Ratio']>0].groupby('Gender').mean()['Avg_Utilization_Ratio'].sort_values(ascending=False))
# we see that average utlization ratio on average is greater for females 
# lets see the spread now 
print(data[data['Avg_Utilization_Ratio']>0].groupby('Gender').std()['Avg_Utilization_Ratio'].sort_values(ascending=False))
# we see that the data is more deviated in the females for the average utilization ratio
#so the ratio is more volatile in the female side as well as higher (does this mean that females use more of the percentage of credit lines )
# lets see the skew
print(data[data['Avg_Utilization_Ratio']>0].groupby('Gender')['Avg_Utilization_Ratio'].skew().sort_values(ascending=False))
# we see that the distribution for males is highly skewed and for females is very lightly skewed. distribution has a longer right tail, more data 
# lets see the outliers through a box plot 
data[data['Gender']=='M'][['Avg_Utilization_Ratio']].boxplot()
plt.pyplot.show()
#distribution has so many data points to the right for male
data[data['Gender']=='F'][['Avg_Utilization_Ratio']].boxplot()
plt.pyplot.show()
# not the case with females, a more approximately normal distribution


#education



#marital status 
```

```python
#income
print(data[data['Avg_Utilization_Ratio']>0].groupby('Income_Category').mean()['Avg_Utilization_Ratio'].sort_values(ascending=False))
# we see that high income groups have a lower utilization ratio on average and as income levels fall the utilization ratio increases
# lets see the spread now 
print(data[data['Avg_Utilization_Ratio']>0].groupby('Income_Category').std()['Avg_Utilization_Ratio'].sort_values(ascending=False))
# we see that the spread gets lower as income levels rise so lower income levels have a greater spread around the mean than higher income groups 
# lets see the skew
print(data[data['Avg_Utilization_Ratio']>0].groupby('Income_Category')['Avg_Utilization_Ratio'].skew().sort_values(ascending=False))
# the data is highly skewed. bigger long tails on the right for all income groups except lesser than 40k and 40k-60k . they are approximately symmetric
# lets see the outliers through a box plot 
data[data['Avg_Utilization_Ratio']>0].groupby('Income_Category')[['Avg_Utilization_Ratio']].boxplot(rot=0, fontsize=12,figsize=((20,20)))
plt.pyplot.show()
# the outliers in the plots(income ones can be seen that are skewing the distribution meaning there is a portion of wealthy people that utilizes the credit very much extended to them)

```

```python
#marital status
print(data.groupby('Marital_Status').mean()['Avg_Utilization_Ratio'].sort_values(ascending=False))
print(data.groupby('Marital_Status').std()['Avg_Utilization_Ratio'].sort_values(ascending=False))
print(data.groupby('Marital_Status')['Avg_Utilization_Ratio'].skew().sort_values(ascending=False))
# lets see the outliers through a box plot 
data.groupby('Marital_Status')[['Avg_Utilization_Ratio']].boxplot(rot=0, fontsize=12,figsize=((20,20)))
plt.pyplot.show()
#the skew tells us that the distributions for all marital groups are positive skewed meaning long right tail so most of them lie on theright side of the curve in terms of mean
# interestingly divorced people have the least utilization ratios meaning they are not using much of the credit lines extended to them 
#as the box plot shows it is not a much skewed in terms of marital groups as there no outliers in the boxplot 
```
